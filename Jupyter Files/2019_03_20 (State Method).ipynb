```python
from abc import ABC, abstractmethod

data = [3, 5, 7, 1, 2, 7, 0]
index = 0

class BaseState:
    currentState = None
    stateList = {}

    @classmethod
    def NextState(cls, name):
        if name in cls.stateList:
            cls.currentState = cls.stateList[name]
            cls.currentState.EnterState()

    def __init__(self, name):
        self.name = name
        BaseState.stateList[name] = self

    @abstractmethod
    def Message(self):
        print(self.name + " State")

    @abstractmethod
    def EnterState(self):
        print("\nEntering " + self.name + " State")
    

class TriangleState(BaseState):
    def __init__(self):
        BaseState.__init__(self, "TriangleState")

    def Message(self):
        global data
        global index
        BaseState.Message(self)
        if data[index] < 5:
            BaseState.NextState("SquareState")
        else:
            BaseState.NextState("CircleState")
        index += 1

    def EnterState(self):
        BaseState.EnterState(self)


class SquareState(BaseState):
    def __init__(self):
        BaseState.__init__(self, "SquareState")

    def Message(self):
        global data
        global index
        BaseState.Message(self)
        if data[index] & 1:
            BaseState.NextState("PentagonState")
        else:
            BaseState.NextState("TriangleState")
        index += 1

    def EnterState(self):
        BaseState.EnterState(self)


class CircleState(BaseState):
    def __init__(self):
        BaseState.__init__(self, "CircleState")

    def Message(self):
        global data
        global index
        BaseState.Message(self)
        if data[index] % 3 == 0:
            BaseState.NextState("SquareState")
        else:
            BaseState.NextState("TriangleState")
        index += 1

    def EnterState(self):
        BaseState.EnterState(self)


class PentagonState(BaseState):
    def __init__(self):
        BaseState.__init__(self, "PentagonState")

    def Message(self):
        global data
        global index
        BaseState.Message(self)
        if data[index] == 1 or data[index] == 5 or data[index] == 9:
            BaseState.NextState("CircleState")
        else:
            BaseState.NextState("PentagonState")
        index += 1

    def EnterState(self):
        BaseState.EnterState(self)

t = TriangleState()
s = SquareState()
c = CircleState()
p = PentagonState()
BaseState.currentState = t

while data[index] != 0:
    BaseState.currentState.Message()
```
