```python
# __len__, __setitem__, __getitem__

import math

class Vector2:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
        
    def _AreNums(self, lst):
        if ((isinstance(lst[0], int) or isinstance(lst[0], float)) and (isinstance(lst[1], int) or isinstance(lst[1], float))):
            return True
        return False
        
    def __add__(self, other):
        if isinstance(other, Vector2):
            return Vector2(self.x + other.x, self.y + other.y)
        return self.__radd__(other)
        
    def __radd__(self, other):
        if ((isinstance(other, tuple) or isinstance(other, list)) and
            len(other) == 2 and self._AreNums(other)):
            return Vector2(self.x + other[0], self.y + other[1])
        return NotImplemented
        
    def __iadd__(self, other):
        if isinstance(other, Vector2):
            self.x += other.x
            self.y += other.y
            return self
        if ((isinstance(other, tuple) or isinstance(other, list)) and
            len(other) == 2 and self._AreNums(other)):
            self.x += other[0]
            self.y += other[1]
            return self
        return NotImplemented
    
    def __eq__(self, other):
        if isinstance(other, Vector2):
            return self.x == other.x and self.y == other.y
        if ((isinstance(other, tuple) or isinstance(other, list)) and
            len(other) == 2 and self._AreNums(other)):
            return self.x == other[0] and self.y == other[1]
        return NotImplemented
    
    def __str__(self):
        return f"<{self.x},{self.y}>"
    
    def Length(self):
        return math.sqrt(self.x * self.x + self.y * self.y)
    
    def __len__(self):
        return 2
    
    def __setitem__(self, key, value):
        if isinstance(key, int) or isinstance(key, float):
            if key < -2 or key > 1 or key != math.floor(key):
                raise IndexError
            if key == 0 or key == -2:
                self.x = value
            else:
                self.y = value
            return
        if isinstance(key, str):
            if key == 'x' or key == 'X':
                self.x = value
            elif key == 'y' or key == 'Y':
                self.y = value
            else:
                raise KeyError
            return
        raise TypeError
    
    def __getitem__(self, key):
        if isinstance(key, int) or isinstance(key, float):
            if key < -2 or key > 1 or key != math.floor(key):
                raise IndexError
            if key == 0 or key == -2:
                return self.x
            return self.y
        if isinstance(key, str):
            if key == 'x' or key == 'X':
                return self.x
            if key == 'y' or key == 'Y':
                return self.y
            raise KeyError
        raise TypeError

v1 = Vector2(1,2)
print(len(v1))
print(v1[0])
print(v1['Y'])
v1[0] = 3
v1[1] = 4
print(v1)
v1['x'] = 5
v1['y'] = 6
print(v1)
for i in v1: #for loop uses the __getitem__ call; knows to stop when IndexError is raised
    print(i)
```
