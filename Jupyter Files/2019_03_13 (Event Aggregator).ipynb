```python
# Event Aggregator

from ui import *

class Event:
    def __init__(self, id):
        self.id = id

class EventManager:
    def __init__(self):
        self.events = []
        
    def PostEvent(self, event):
        self.events.append(event)

    def __iter__(self):
        return self
    
    def __next__(self):
        if len(self.events) == 0:
            raise StopIteration
        return self.events.pop(0)

eventManager = EventManager()
    
class Independent:
    a = 10
    def Add(self, tmp):
        global eventManager
        Independent.a += 1
        eventManager.PostEvent(Event("Add"))

class Dependent1:
    def Update(self):
        self.a = Independent.a
        print("New Number: " + str(self.a))

class Dependent2:
    def __init__(self):
        self.a = Independent.a
    
    def Update(self):
        print("Last Number: " + str(self.a))
        self.a = Independent.a

i = Independent();

root = Tk()
f = UIFrame(root=root)
b = UIButton(i.Add, text="Add", bg="green")
f.Add(b, side=LEFT)

d1 = Dependent1()
d2 = Dependent2()

def Update():
    global eventManager
    for i in eventManager:
        if i.id == "Add":
            d1.Update()
            d2.Update()
    root.after(10, Update)

root.after(10, Update)
root.mainloop()
```
