```python
# Mediator Pattern

from ui import *

class MessageSystem:
    def __init__(self):
        self.messageCallbacks = {}
        
    def RegisterTarget(self, name):
        self.messageCallbacks[name] = []
        
    def PostMessage(self, name, message):
        self.messageCallbacks[name].append(message)
    
    def GetMessages(self, name):
        lst = self.messageCallbacks[name].copy()
        self.messageCallbacks[name].clear()
        return lst

messageSystem = MessageSystem()

class Independent:
    a = 10
    def Add(self, tmp):
        global messageSystem
        Independent.a += 1
        messageSystem.PostMessage("d1", "Add")
        messageSystem.PostMessage("d2", "Add")

class Dependent1:
    def __init__(self, name):
        global messageSystem
        self.name = name
        messageSystem.RegisterTarget(name)
        
    def Update(self):
        global messageSystem
        messages = messageSystem.GetMessages(self.name)
        for m in messages:
            if m == "Add":
                self.a = Independent.a
                print("New Number: " + str(self.a))

class Dependent2:
    def __init__(self, name):
        global messageSystem
        self.a = Independent.a
        self.name = name
        messageSystem.RegisterTarget(name)
    
    def Update(self):
        global messageSystem
        messages = messageSystem.GetMessages(self.name)
        for m in messages:
            if m == "Add":
                print("Last Number: " + str(self.a))
                self.a = Independent.a

i = Independent();

root = Tk()
f = UIFrame(root=root)
b = UIButton(i.Add, text="Add", bg="green")
f.Add(b, side=LEFT)

d1 = Dependent1("d1")
d2 = Dependent2("d2")

def Update():
    d1.Update()
    d2.Update()
    root.after(10, Update)

root.after(10, Update)
root.mainloop()
```
