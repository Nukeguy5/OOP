```python
# Inline 'if' statement

a = 10
b = 20
c = 20

if a == b:
    print('a == b')
else:
    print('a != b')
    
if b == c:
    print('b == c')
else:
    print('b != c')

print()

print('a == b' if a == b else 'a != b')
print('b == c' if b == c else 'b != c')
```

```python
# Inline 'for' loop - for creating a list

lst = [1, 2, 3, 10, 5]

print(*lst)  # '*' passes each item in the list as a parameter (leaves the '[]' out)
print()
print(*[str(i + 1) for i in lst])
```

```python
# With generators we are using 'tuples' for comprehension instead of 'list' or 'dict'

# Find all words in the dictionary that begin with "chroma"

with open("words_alpha.txt") as infile:
    chroma = (w for w in infile if w.startswith("chroma"))  # does not read the file at this time
#   w = Each line in 'infile'
#   () indicates that it is a generator

    for i in chroma:  # <- reads the file here
        print(i)
```

```python
# Custom generator

def words_begin_chroma(file):
    for w in file:
        if w.startswith("chroma"):
            yield w  # 'yield' mean exit here and next time this function is called start here
            
with open("words_alpha.txt") as infile:
    chroma = words_begin_chroma(infile)  # creates a custom generator out of this because of 'yield'
    for i in chroma:
        print(i)
```
